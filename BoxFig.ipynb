```python
import src.BoundingPolygon as BoundingPolygon
import numpy as np
import matplotlib.pyplot as plt


pts = np.random.rand(1000, 2) * 9  # 1000 random points in the range [0, 9]
bp = BoundingPolygon.BoundingPolygon([(3, 3), (3, 6), (6, 6), (6, 3), (3, 3)])
inside = bp.inPolygon(pts)

# Set color red if in box, blue if outside box
color = np.empty(len(inside), dtype=str)
color[inside] = (
    "k"  ### changed to black just to write a long comment that we can then lint to the right length, let's see how this goes blah blah blah
)
color[~inside] = "b"


# Plot
fig, ax = plt.subplots(figsize=(6, 6))
ax.scatter(pts[:, 0], pts[:, 1], c=color)
```
